```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

```

## Functions

```python
def distribution_plot (x):
    #nbins = int(x.max())-int(x.min())+1
    #print (type(nbins))
    plt.figure(figsize=(10, 8))
    ax = sns.distplot(x, )
    
    # Auxiliary information:
    mn = x.mean()
    mx = ax.lines[0].get_ydata().max()
    
    # Plot median line:
    ax.plot([mn]*2, [0, mx])
    
    
    sns.set(style='white', palette='muted', color_codes=True)
    plt.annotate('Mean', [mn, mx], xytext=[mn*1.1, mx*1.1], fontsize=10, arrowprops=dict(arrowstyle='->', color='red',connectionstyle='arc3, rad=.2'))
    plt.show()
```

```python
# Reading the data
file_path = '/home/mav24/Documents/Development/Regeneration/Project/Data/mpg.data.xlsx'
data=pd.read_excel(file_path)
```

```python
# Droping Empty columns
data.drop(columns=['Unnamed: 9', 'Unnamed: 10', 'Unnamed: 11', 'Unnamed: 12'], inplace=True)
data.columns=['mpg', 'cylinders', 'displacements', 'horsepower', 'weight', 'acceleration', 'model year', 'origin', 'car name']
# Sanity check. Count of all the data (rows)
print (f'Total number of data: {len(data)}')

# Dropping rows with na 
data.dropna(axis=0, inplace=True)
print (f'Total number of data after dropping NaN values: {len(data)}')# Sanity check. Count of all the data (rows)
print (f'Shape of our data{data.shape}')
```

```python
# Set figure size
plt.figure(figsize=(10, 8))

# Barplot cylinders and displacements
sequential_colors = sns.color_palette("OrRd", 5)
sns.barplot(x=data['cylinders'], y=data['displacements'], data=data, palette=sequential_colors)
plt.xlabel('Cylinders')
plt.ylabel('Displacements')
plt.title('Displacements by number of Cylinders')
plt.show()
```

```python
# Set figure size
plt.figure(figsize=(10, 8))

# Barplot cylinders and displacements
diverging_colors = sns.color_palette("OrRd", 13)
sns.barplot(x=data['model year'], y=data['mpg'], data=data, palette=diverging_colors)
plt.xlabel('Model Year')
plt.ylabel('MPG')
plt.title('MPG by Year of the model of the car')
plt.show()
```

```python
# Correlation map IMPORTANT

# Create correlation matrix for the data
corr = data.corr()

# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(240, 10, n=9)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=1, vmin=-1, center=0, annot=True,
            square=True, linewidths=.5, cbar_kws={"shrink": 0.5})
```

```python
# Set figure size
plt.figure(figsize=(10, 8))

# MPG Diplacements polynomial model 2nd order
sns.regplot(x=data['displacements'], y=data['mpg'], data=data, order=2,
           line_kws={'color':'red'})
plt.xlabel('Displacements')
plt.ylabel('MPG')
plt.title('Scatter plot of MPG and Displacements')
plt.show()
print ('Second order polynomial model fit the correlation between MPG and Displacements')
```

```python
# Set figure size
plt.figure(figsize=(10, 8))

# MPG Horsepower polynomial model 2nd order
sns.regplot(x=data['horsepower'], y=data['mpg'], data=data, order=2,
           line_kws={'color':'red'})
plt.xlabel('Horsepower')
plt.ylabel('MPG')
plt.title('Scatter plot of MPG and Horsepower')
plt.show()
print ('Second order polynomial model fit the correlation between MPG and Horsepower')
```
